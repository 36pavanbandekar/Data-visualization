```python
# # DATA VISUALIZATION USING seaborn
# Import the python libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets
iris = pd.read_csv('Iris.csv')
print(iris.head())

wine_reviews = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
wine_reviews.head()
```

```python
# Scatter plot
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',color="red", hue="Species", data=iris)
plt.show()
```

```python
sns.relplot(
    data=iris,
    x='SepalLengthCm', y='SepalWidthCm', hue="Species"
)
```

```python
# Line chart
sns.lineplot(data=iris.drop(['Id','Species'], axis=1))
plt.show()
```

```python
# Bar chart
sns.countplot('Species',data=iris)
plt.show()
```

```python
# Histogram
sns.histplot(wine_reviews['points'], bins=30)
plt.show()
```

```python
# Box plot 
df = wine_reviews[(wine_reviews['points']>=95) & (wine_reviews['price']<1000)]
sns.boxplot('points', 'price', data=df)
plt.show()
```

```python
iris.corr()
```

```python
# Heat map
cmap = sns.dark_palette('#3fdd01', as_cmap=True)
sns.heatmap(iris.corr(), annot=True, vmin =-1, vmax=1,  cmap=cmap)
plt.show()
```

```python
cmap= sns.cubehelix_palette()
sns.heatmap(iris.corr(), vmin =-1, vmax=1, annot=True, cmap=cmap)
```

```python
cmap = sns.diverging_palette(320, 40, sep=50, as_cmap=True)
sns.heatmap(iris.corr(),annot=True,cmap=cmap)
```

```python
# Example 1
# Scatter plot with color distinction
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris) 
plt.show()
```

```python
# Example 2
# Histogram with kernel density estimate

sns.histplot(wine_reviews['points'], bins=10, kde=True)
plt.show()
```

```python
# Example 3
# Faceting 
g = sns.FacetGrid(iris, col='Species')
g = g.map(sns.kdeplot, 'SepalLengthCm')
plt.show()
```

```python
# Example 4
# Joint plot

sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, height=5, 
              kind="reg")
plt.show()
```

```python
# Example 5
# Pair plot

sns.pairplot(iris.drop("Id", axis=1), hue="Species", height=3, diag_kind="hist")
plt.show()
```

```python
# Example 6
# Violin plot

sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
plt.show()
```
